```python
x = np.array(list(range(8)))
y = np.array(list(range(8)))**2
z = np.array(list(range(8)))**3
```

```python
data = {'x1': x, 'x2':y, 'x3': z}
```

```python
my_df = pd.DataFrame(data)
```

```python
my_df
```

```python
my_df.iplot(x = 'x1', y = 'x2', mode = 'markers')
```
